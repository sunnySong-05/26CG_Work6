```python
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output 

from pytorch3d.io import load_obj, save_obj
from pytorch3d.structures import Meshes
from pytorch3d.utils import ico_sphere
from pytorch3d.loss import mesh_edge_loss, mesh_laplacian_smoothing, mesh_normal_consistency
from pytorch3d.renderer import (
    look_at_view_transform, FoVPerspectiveCameras,
    RasterizationSettings, MeshRasterizer, SoftSilhouetteShader, BlendParams
)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# 1. 文件夹准备
output_img_dir = "cow_evolution_long"
os.makedirs(output_img_dir, exist_ok=True)

# 2. 目标数据准备
obj_path = "cow.obj"
verts, faces, _ = load_obj(obj_path)
verts = verts.to(device)
faces_idx = faces.verts_idx.to(device)
verts = (verts - verts.mean(0)) / verts.abs().max() 
target_mesh = Meshes(verts=[verts], faces=[faces_idx])

# 3. 渲染器配置
num_views = 20
R, T = look_at_view_transform(dist=2.7, elev=torch.linspace(0, 360, num_views), azim=torch.linspace(-180, 180, num_views))
cameras = FoVPerspectiveCameras(device=device, R=R, T=T)
rasterizer = MeshRasterizer(cameras=cameras, raster_settings=RasterizationSettings(image_size=128, blur_radius=1e-4, faces_per_pixel=100))
shader = SoftSilhouetteShader(blend_params=BlendParams(sigma=1e-4, gamma=1e-4))

with torch.no_grad():
    target_silhouette = shader(rasterizer(target_mesh.extend(num_views)), target_mesh.extend(num_views))[..., 3]

# 4. 初始化源模型与优化器
src_mesh = ico_sphere(4, device)
deform_verts = torch.zeros(src_mesh.verts_packed().shape, device=device, requires_grad=True)

# 初始学习率设为 0.02，兼顾形变速度与平滑度
optimizer = torch.optim.Adam([deform_verts], lr=0.02)
# 每 200 步将学习率减半，确保后期能收敛到细节
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=200, gamma=0.5)

# 5. 优化循环 (拉长到 600 步)
epochs = 600
for i in range(epochs):
    optimizer.zero_grad()
    new_src_mesh = src_mesh.offset_verts(deform_verts)
    pred_silhouette = shader(rasterizer(new_src_mesh.extend(num_views)), new_src_mesh.extend(num_views))[..., 3]
    
    loss_sil = ((pred_silhouette - target_silhouette) ** 2).mean() * 200.0
    
    # 动态正则化：前期为了让它“跑得动”，正则化给小一点；后期为了“长得顺”，正则化给大一点
    w_reg = 0.5 if i < 300 else 1.5
    loss = loss_sil + w_reg * (mesh_laplacian_smoothing(new_src_mesh) + mesh_edge_loss(new_src_mesh)) + 0.1 * mesh_normal_consistency(new_src_mesh)
    
    loss.backward()
    optimizer.step()
    scheduler.step()

    if i % 20 == 0 or i == epochs - 1:
        clear_output(wait=True)
        fig, ax = plt.subplots(1, 2, figsize=(10, 5))
        ax[0].imshow(target_silhouette[0].cpu().numpy(), cmap='gray')
        ax[0].set_title("Target")
        ax[1].imshow(pred_silhouette[0].detach().cpu().numpy(), cmap='gray')
        ax[1].set_title(f"Iteration {i} (LR: {optimizer.param_groups[0]['lr']:.4f})")
        plt.savefig(os.path.join(output_img_dir, f"step_{i:03d}.png"))
        plt.show()
        print(f"进度: {i:03d}/{epochs} | Loss: {loss.item():.4f}")

save_obj("evolution_final.obj", new_src_mesh.verts_list()[0], new_src_mesh.faces_list()[0])
```
